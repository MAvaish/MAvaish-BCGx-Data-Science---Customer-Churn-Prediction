```python
import pandas as pd

price_data = pd.read_csv(r"C:\Users\91739\Downloads\price_data (1).csv")
client_data = pd.read_csv(r"C:\Users\91739\Downloads\client_data (1).csv")
print(price_data.head())
print(client_data.head())
```

```python
# Data types of each column
print("\nüîç Price Data - Column Data Types:")
print(price_data.dtypes)

print("\nüîç Client Data - Column Data Types:")
print(client_data.dtypes)
```

```python
# Numerical summary
print("\nüìä Price Data - Descriptive Statistics:")
print(price_data.describe())

print("\nüìä Client Data - Descriptive Statistics:")
print(client_data.describe())

# Categorical column analysis
print("\nüîπ Price Data - Unique Values Count:")
print(price_data.nunique())

print("\nüîπ Client Data - Unique Values Count:")
print(client_data.nunique())
```

```python
# Churn rate analysis
churn_rate = client_data['churn'].mean()
print(f"\nüöÄ Client Churn Rate: {churn_rate:.2%}")

# Checking unique IDs
unique_price_ids = price_data['id'].nunique()
unique_client_ids = client_data['id'].nunique()
print(f"üîπ Unique IDs in Price Data: {unique_price_ids}")
print(f"üîπ Unique IDs in Client Data: {unique_client_ids}")
```
