```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

price_data = pd.read_csv(r"C:\Users\91739\Downloads\price_data (1).csv")
client_data = pd.read_csv(r"C:\Users\91739\Downloads\client_data (1).csv")

print(price_data.head())
print(client_data.head())
```

```python
num_cols_price = price_data.select_dtypes(include=['float64', 'int64']).columns.tolist()
num_cols_client = client_data.select_dtypes(include=['float64', 'int64']).columns.tolist()

print("ðŸ“Š Numerical Columns in Price Data:", num_cols_price)
print("ðŸ“Š Numerical Columns in Client Data:", num_cols_client)


```

```python
def plot_distribution(data, column_name):
    """
    Function to plot histogram and box plot for a given column.
    """
    plt.figure(figsize=(12, 5))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(data[column_name], bins=30, kde=True, color='blue')
    plt.title(f"Histogram of {column_name}")

    # Box Plot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=data[column_name], color='orange')
    plt.title(f"Box Plot of {column_name}")

    plt.show()
```

```python
# Loop through numerical columns in price_data and plot distributions
for col in num_cols_price:
    print(f"ðŸ“Š Distribution of {col}:")
    plot_distribution(price_data, col)
```

```python
# Loop through numerical columns in client_data and plot distributions
for col in num_cols_client:
    print(f"ðŸ“Š Distribution of {col}:")
    plot_distribution(client_data, col)
```
